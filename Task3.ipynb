STEP 1:: DATA PRE_PROCESSING AND ANALYSIS

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn import model_selection
from sklearn.model_selection import train_test_split
```

```python
dataset=pd.read_csv('/content/drive/MyDrive/data/Housing.csv')
```

```python
data=pd.get_dummies(dataset)
```

```python
data
```

STEP 2:: DATA SPLITTING

```python
x=data.drop('price',axis=1)
y=data['price']
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

STEP 3:: MODEL TRAINING

```python
from sklearn.linear_model import LinearRegression
```

```python
equation=LinearRegression()
equation.fit(x_train,y_train)
```

```python
equation.intercept_
```

```python
equation.coef_
```

```python
y_test_predicted=equation.predict(x_test)
y_test_predicted
```

```python
sns.regplot(x=y_test,y=y_test_predicted)
```

STEP 4:: MODEL EVALUATION

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

r2 = r2_score(y_test, y_test_predicted)
mae = mean_absolute_error(y_test, y_test_predicted)
mse = mean_squared_error(y_test, y_test_predicted)
rmse = np.sqrt(mse)
```

```python
print(f"R-squared: {r2}")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
```
